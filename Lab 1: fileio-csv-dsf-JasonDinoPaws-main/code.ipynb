# letter Frequency

```python
LF = {}

with open("72756.txt.utf-8") as file:
    # Goes though all lines and removes the whitespace
    for Li in file.readlines():
        word = Li.strip()

        # Goes to get all of the Letters within the line
        for P in range(len(word)):
            Le = word[P]

            # If letter and letter not space
            if Le and Le != " ":
                # If already add else Create new
                if Le in LF:
                    LF[Le] += 1
                else: 
                    LF[Le] = 1

# Sorts the items based off of the value  from acending - decending
for L in sorted(LF.items(),key = lambda x:x[1], reverse = True):
    print(f"{L[0]}: {L[1]}")
```

# Break this Cipher

```python
Chipher = input("Chipher: ")
alpha = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']

# shifts for 26 letters
for s in range(26):
    text = ""

    # goes thought all chars in the Chipher
    for p in range(len(Chipher)):
        L = Chipher[p]
        
        # if letter is in aplhabet shifts the lowered version
        if L.lower() in alpha:
            ap = alpha.index(L.lower())
            L = alpha[ap-(s+1)]
        
        # if orginal was captial sets to to capital
        if L.isupper():
            text += L.upper()
        else:
            text += L

    # For ever shift print the amount and the new string
    print("-"*50)
    print(f"Shift: {s+1}\nChipher: {text}")
```

# Log Files

```python
ips = []
websites = {}
textrepot = "\n\n"

with open("server_log.txt") as file:
    for line in file.readlines():
        # Splits by "-" and then formats it in IP, Data, Address
        ip,date,add = (line.strip()).split("-",3)
        textrepot += f"{ip}, {date}, {add}\n"

        if not ip in ips:
            ips.append(ip)

        if add in websites:
            websites[add] += 1
        else:
            websites[add] = 0

with open("server_summary.txt","w") as report:
    report.write(f"Unique Ips: {len(ips)}\n\n") # prints count of Uni Ips

    report.write("Top 3 Sites\n")
    c = 0
    # sorts it from Value accending if the prints isnt grater than 3
    for L in sorted(websites.items(),key = lambda x:x[1], reverse = True):
        if c < 3:
            c += 1
            report.write(f"{L[0]}: {L[1]}\n")
    
    report.write(textrepot)
```

# Receiptify

```python
from tabulate import tabulate
amt = {}

with open("spotify_data.csv") as file:
    Lines = file.readlines()
    Lines.pop(0)
    for line in Lines:
        # Song, Doesnt matter...
        s,a,d,t = line.strip().split(",",4)

        # if there add if not do not
        if s in amt:
            amt[s] += 1
        else:
            amt[s] = 1


# Set up
print('_'*31)
print("|       NOT RECEIPTIFY        |")
print("|          Last Week          |")
print(f"|{'_'*29}|",)
print("|QTY|ITEM                 |AMT|",)
print(f"|{'-'*29}|",)

# For numbers to Format into 000
def calnum(num):
    return num < 10 and f'00{num}' or num < 100 and f'0{num}' or num

c = 1
# Sorts Accedning value up to 10
for L in sorted(amt.items(),key = lambda x:x[1], reverse = True):
    if c < 11:
        # |Place|Song name Up to 21 chars|Count|
        print(f"|{calnum(c)}|{L[0]}{' '*(21-len(L[0]))}|{calnum(L[1])}|")
        c += 1

print('-'*31)
```
